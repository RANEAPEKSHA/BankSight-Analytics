```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/Banking.csv')
df.head(5)
```

```python
#generating descriptive statistic for the dataframe
df.describe()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
#Examin the distribution of unique categories in categories in categorical columns
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification","Properties Owned", "Risk Weighting", "Income Band"]].columns

for col in categorical_cols:
  print(f"value Counts for '{col}':")
  display(df[col].value_counts())
```

# Bivariate Analysis

```python
#bivariate Analysis
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification","Properties Owned", "Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
```

#univariate Analysis

```python

for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification","Properties Owned", "Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

```python
#histplot of value counts for different Occupation

for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

# Numerical Analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']
# Univariate analysis and visualization
plt.figure(figsize=(19, 19))
for i, col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.show()
```

# Heatmaps

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```

# **Quick EDA Insights:**
Thereâ€™s a strong link between bank deposits, checking accounts, and savings
accounts, suggesting that customers who use one often use the others too.

Estimated income shows a moderate relationship with superannuation savings â€” people earning more tend to save more for retirement.

Business lending has a mild connection with bank loans, which makes sense as businesses taking loans often access other lending services.

Superannuation and foreign currency accounts donâ€™t show much correlation with other features, indicating theyâ€™re used more independently.
